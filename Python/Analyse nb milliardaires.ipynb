## Nettoyage CSV 1

```python

# on charge les données de notre premier csv dans la variable data
data = pd.read_csv('/content/drive/MyDrive/Python/Billionaires Statistics Dataset.csv')

# on créer un fonction qui va nous servir a analyser tout les attributs car on souhaite avoir le nombre de valeur null et les supprimer
def analyses_attributs(nom_att):
    cpt_null = data[nom_att].isnull().sum()

    # nombre total de ligne
    ligne_total = len(data)
    # pourcentage de valeur null
    pourcent_null = (cpt_null / ligne_total) * 100

    return {
        "Attribut": nom_att,
        "Nombre de valeurs nulles": cpt_null,
        "Pourcentage de valeurs nulles par rapport au nombre total de lignes": "{:.2f}%".format(pourcent_null),
    }

# la liste des attributs selectionné que l'on souhaite analyser
attributs_selec = [
    'finalWorth', 'category', 'personName', 'age', 'country', 'industries',
    'countryOfCitizenship', 'selfMade', 'status', 'gender', 'cpi_country', 'cpi_change_country', 'gdp_country',
    'gross_tertiary_education_enrollment', 'gross_primary_education_enrollment_country',
    'life_expectancy_country', 'tax_revenue_country_country', 'total_tax_rate_country',
    'population_country'
]

# on créer un tableau avec nos resultats
tableau_resultats = []

# boucle qui va parcourir tout les attributs
for attribut in attributs_selec:
    tableau_resultats.append(analyses_attributs(attribut))

# on converti le tableau en dataframe
resultats_df = pd.DataFrame(tableau_resultats)

# on va afficher notre dataframe avec tabulate oarce que c'est bien plus beau comme ça
print(tabulate(resultats_df, headers='keys', tablefmt='fancy_grid'))
```

```python
# on charge les données de notre second csv dans la variable data
data = pd.read_csv('/content/drive/MyDrive/Python/Billionaires Statistics Dataset.csv')

# on créer un fonction analyser tout les attributs et supprimer les valeurs nulles des attributs spécifique
def recup_attribut(nom_att, dataframe):
    cleaned_data = dataframe.dropna(subset=[nom_att])

    # Calculer le nombre de valeurs nulles dans la colonne après suppression des lignes
    cpt_null_supp = dataframe.shape[0] - cleaned_data.shape[0]

    return cleaned_data, cpt_null_supp

# on copie nos données pour éviter de modifier le csv
cleaned_data = data.copy()
ligne_total = len(cleaned_data)

# var qui stock le nombre de ligne supp
ligne_supp = 0

# Analyser chaque attribut et supprimer les lignes avec des valeurs nulles
for attribut in attributs_selec:
    cleaned_data, cpt = recup_attribut(attribut, cleaned_data)
    ligne_supp += cpt

# Calculer le nombre total de lignes restantes
ligne_restante = ligne_total - ligne_supp

# Calculer le pourcentage de lignes supprimées par rapport au nombre total de lignes
pourcent_supp = (ligne_supp / ligne_total) * 100

# Afficher les résultats
print("Nombre total de lignes dans le fichier:", ligne_total)
print("Nombre total de lignes supprimées:", ligne_supp)
print("Pourcentage de lignes supprimées par rapport au nombre total de lignes: {:.2f}%".format(pourcent_supp))
print("Nombre total de lignes restantes après suppression:", ligne_restante)
print("Nombre total de lignes restantes après suppression:", len(cleaned_data))
```

```python
# on compte le nombre de val pour chaque catégorie et on calcul leur pourcentage
cpt_statut = cleaned_data['status'].value_counts()
ligne_restante = len(cleaned_data)

print("Nombre total de lignes restantes après suppression:", ligne_restante)

# Afficher le nombre et le pourcentage de chaque catégorie
for statut, cpt in cpt_statut.items():
    pourcentage = (cpt / ligne_restante) * 100
    print("Statut:", statut)
    print("Nombre total:", cpt)
    print("Pourcentage: {:.2f}%".format(pourcentage))
```

```python
# les nombres de lignes avant la suppression
total_ligne = len(cleaned_data)
print("Nombre total de lignes avant suppression:", total_ligne)

# les valeurs que l'on supprime car on ne connait pas leur signification
a_supp = ['E', 'N', 'R']

# on supprime les lignes qui ont les val définit dans la variable a_supp
cleaned_data_st = cleaned_data[~cleaned_data['status'].isin(a_supp)]

# on recup le nombre de ligne total après modification
total_ligne_v2 = len(cleaned_data_st)
print("Nombre total de lignes après suppression:", total_ligne_v2)

# et on définit une valeur qui aura le nombre de lignes supprimées
ligne_supp = total_ligne - total_ligne_v2
print("Nombre de lignes supprimées:", ligne_supp)

# et on donne le pourcentage de lignes supprimées
pourcent_supp = (ligne_supp / total_ligne) * 100
print("Pourcentage de lignes supprimées: {:.2f}%".format(pourcent_supp))
```

```python
# Fonction pour analyser chaque attribut et supprimer les lignes avec des valeurs nulles
def analyze_attribute(attribute_name, dataframe):
    # Supprimer les lignes avec des valeurs nulles pour cet attribut
    cleaned_data_v2 = dataframe.dropna(subset=[attribute_name])

    # Calculer le nombre de valeurs nulles dans la colonne après suppression des lignes
    cpt_null_2 = dataframe.shape[0] - cleaned_data_v2.shape[0]

    return cleaned_data_v2, cpt_null_2

# Copie des données pour éviter de modifier les données d'origine
cleaned_data_v2 = data_1.copy()
ligne_total = len(cleaned_data_v2)

# Initialisation du nombre total de lignes supprimées
ligne_supp = 0

# Analyser chaque attribut et supprimer les lignes avec des valeurs nulles
for attribute in attributs_selec:
    cleaned_data_v2, _cpt_null = analyze_attribute(attribute, cleaned_data_v2)
    ligne_supp += _cpt_null

# Calculer le nombre total de lignes restantes
ligne_restante = ligne_total - ligne_supp

# Calculer le pourcentage de lignes supprimées par rapport au nombre total de lignes
pourcentage_supp = (ligne_supp / ligne_total) * 100

# Afficher les résultats
print("Nombre total de lignes dans le fichier:", ligne_total)
print("Nombre total de lignes supprimées:", ligne_supp)
print("Pourcentage de lignes supprimées par rapport au nombre total de lignes: {:.2f}%".format(pourcentage_supp))
print("Nombre total de lignes restantes après suppression:", ligne_restante)
```

# Code

Pour notre première analyse nous allons commencer par l'attribut catégorie

```python
# Compter le nombre de personnes dans chaque catégorie
cpt_personne = cleaned_data['category'].value_counts().reset_index()
cpt_personne.columns = ['Catégorie', 'Nombre de personnes']

# Créer le graphique à barres avec Plotly Express
fig = px.bar(cpt_personne, x='Catégorie', y='Nombre de personnes',
             title='Nombre de milliardaires dans chaque catégorie',
             labels={'Catégorie': 'Catégorie', 'Nombre de personnes': 'Nombre de personnes'})
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

Ce premier graphique nous montre la répartition du nombre de milliardaire par catégorie, on peut voir qu'il y a une forte concentration dans certain secteur.

Nous observons que la catégorie des finances et les investissements est la plus représentée, suivie de près par la manufacturing (fabrication) et le secteur de la technologie.

Les résultats mettent en lumière la diversité des industries dans lesquelles les milliardaires opèrent, reflétant ainsi les différentes voies vers la richesse et le succès financier.





```python
# Filtrer les données pour les années de 2001 à 2023
cleaned_data_v2 = cleaned_data_v2[(cleaned_data_v2['year'] >= 2001) & (cleaned_data_v2['year'] <= 2023)]

# Enlever "B" et convertir net_worth en int
cleaned_data_v2['net_worth'] = cleaned_data_v2['net_worth'].str.replace(' B', '').str.strip().astype(float)

# Créer le graphique animé avec une animation_frame
fig = px.scatter(cleaned_data_v2, x='age', y='net_worth', animation_frame='year',
                 hover_name='full_name', color='gender',
                 title='Évolution du rang et de la valeur nette au fil des ans')

# Durée de la transition
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 1500

fig.update_layout(xaxis_title='Age', yaxis_title='Valeur nette (en milliards de dollars)')
fig.show()
```

Ce graphique animé offre une visualisation de l'évolution du rang et de la fortune des milliardaires de 2001 à 2023. Les points sont colorés en fonction du sexe des milliardaires, offrant ainsi une dimension supplémentaire à l'analyse. Le graphique va nous permettre d'avoir une visualisation temporelle de la population.

En observant la progression chronologique à travers les années, nous pouvons identifier des tendances et des motifs dans la distribution de la richesse parmi les milliardaires. Effectivement, on peut aisément identifier qu'il y a eu un accroissement du nombre de milliardaires au fil des années, tout comme une augmentation de leur richesse.

Autre fait notable que l'on peut voir facilement sur ce graphique : l'augmentation du nombre de femmes milliardaires sur ces 20 dernières années alors qu'elles n'étaient que 2 en 2001 !

```python
# Pivoter les données pour obtenir le décompte des genres par catégorie
genre_counts = cleaned_data.pivot_table(index='category', columns='gender', aggfunc='size', fill_value=0).reset_index()

# Remodeler les données pour Plotly Express
genre_counts_melted = pd.melt(genre_counts, id_vars='category', var_name='Genre', value_name='Nombre de personnes')

# Créer le graphique à barres empilées avec Plotly Express
fig = px.bar(genre_counts_melted, x='category', y='Nombre de personnes', color='Genre',
             title='Répartition des genres par catégorie',
             labels={'category': 'Catégorie', 'Nombre de personnes': 'Nombre de personnes', 'Genre': 'Genre'},
             barmode='stack')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

Ce graphique à barres empilées offre une représentation visuelle de la répartition des genres au sein de différentes catégories de milliardaires. Chaque barre représente une catégorie spécifique, tandis que les segments colorés à l'intérieur de chaque barre représentent la répartition des genres (masculin et féminin) au sein de cette catégorie.

En analysant les résultats, nous observons des variations significatives dans la composition de genre selon les catégories. Par exemple, dans certaines catégories telles que les technologies et les finances, la présence masculine est prédominante, tandis que dans d'autres catégories comme la mode et le manufacturing, nous observons que les femmes sont plus representées.

ANALYSE COUNTRY

```python
# On calcule le nombre de milliardaires par pays
nombre_billionaires_par_pays = cleaned_data['country'].value_counts()

# On définit un seuil à 1.5% pour que le graphique soit lisible
seuil = 0.015 * nombre_billionaires_par_pays.sum()

# on identifie et regroupe les pays en dessous du seuil
pays_petits = nombre_billionaires_par_pays[nombre_billionaires_par_pays < seuil]
nombre_billionaires_par_pays['Autres'] = pays_petits.sum()
nombre_billionaires_par_pays = nombre_billionaires_par_pays[nombre_billionaires_par_pays >= seuil]

# on créé un DataFrame avec les pays et leurs nombres de milliardaires
donnees = {'pays': nombre_billionaires_par_pays.index, 'nombre': nombre_billionaires_par_pays.values}
df = pd.DataFrame(donnees)

# On créé le graphique
fig = px.pie(df, values='nombre', names='pays', title='Répartition des milliardaires par pays',
             labels={'nombre':'Nombre de milliardaires', 'pays':'Pays'},
             hole=0.4)
fig.show()

# On calcule la somme des fortunes par pays
richesse_par_pays = cleaned_data.groupby('country')['finalWorth'].sum()

# On définit un seuil à 1% pour que le graphique soit lisible
seuil = 0.01 * richesse_par_pays.sum()

# on identifie et regroupe les pays en dessous du seuil
petits_pays = richesse_par_pays[richesse_par_pays < seuil]
richesse_par_pays['Autres'] = petits_pays.sum()
richesse_par_pays = richesse_par_pays[richesse_par_pays >= seuil]

# on créé un DataFrame avec les pays et la fortune
donnees = {'pays': richesse_par_pays.index, 'richesse_totale': richesse_par_pays.values}
df = pd.DataFrame(donnees)

# On créé le graphique
fig = px.pie(df, values='richesse_totale', names='pays', title='Répartition des fortunes par pays',
             labels={'richesse_totale':'Fortune totale ($)', 'pays':'Pays'},
             hole=0.4)
fig.show()
```

Les graphiques présentés ci-dessus offrent un aperçu éclairant de la répartition des milliardaires et de leurs fortunes à travers différents pays, révélant des tendances significatives dans la dynamique économique mondiale.

Disparité économique entre pays : Ces graphiques mettent en lumière la disparité économique entre les pays.
```python
# On calcule le nombre de milliardaires par pays
nombre_billionaires_par_pays = cleaned_data['country'].value_counts().reset_index()
nombre_billionaires_par_pays.columns = ['country', 'nombre_billionaires']

# On créé la carte du monde du nombre de milliardaires
fig = px.choropleth(nombre_billionaires_par_pays, locations='country', locationmode='country names', color='nombre_billionaires',
                    hover_name='country', color_continuous_scale=['white','yellow','green','blue','purple','orange','red'],
                    title='Nombre de milliardaires par pays')
fig.update_layout(coloraxis_colorbar=dict(title='Nombre de milliardaires par pays'))
fig.show()

cleaned_data['gdp_country'] = cleaned_data['gdp_country'].replace('[\$,]', '', regex=True).astype(float)

#  On récupere le PIB des différents pays (GDP = PIB)
pib_moyen_par_pays = cleaned_data.groupby('country')['gdp_country'].mean().reset_index()

# On créé la carte du monde avec le PIB
fig = px.choropleth(pib_moyen_par_pays, locations='country', locationmode='country names', color='gdp_country',
                    hover_name='country', color_continuous_scale=['white','yellow','green','blue','purple','orange','red'],
                    title='PIB moyen par pays')
fig.update_layout(coloraxis_colorbar=dict(title='PIB moyen par pays'))
fig.show()
```

Les cartes présentées ci-dessus fournissent des informations supplémentaires sur la dynamique économique mondiale, en mettant en évidence une corrélation notable entre le nombre de milliardaires par pays et le PIB moyen par pays.

Similarités dans les résultats : Les deux cartes révèlent des résultats assez similaires, avec une répartition géographique qui semble se recouper dans une certaine mesure. Les pays qui affichent un nombre élevé de milliardaires tendent également à avoir un PIB moyen relativement élevé, tandis que les pays avec un nombre plus faible de milliardaires ont tendance à afficher un PIB moyen plus bas. Cette observation suggère une corrélation entre la concentration de la richesse et la performance économique globale d'un pays.

Celle-ci peut être interprétée de plusieurs façons. Tout d'abord, un PIB plus élevé peut fournir un environnement économique favorable à la création et à l'accumulation de richesses, ce qui peut conduire à un plus grand nombre de milliardaires. En revanche, un nombre élevé de milliardaires peut également contribuer à stimuler l'économie d'un pays en investissant dans des projets innovants, en créant des emplois et en stimulant la croissance économique.

Facteurs de corrélation : Cette corrélation peut être influencée par divers facteurs, tels que la stabilité politique, le système économique, les politiques fiscales, le niveau d'éducation, l'infrastructure économique, et d'autres indicateurs socio-économiques. Les pays qui réussissent à créer un environnement favorable à l'entrepreneuriat et à l'innovation ont tendance à attirer des investissements et à favoriser la croissance économique, ce qui peut se traduire par une augmentation du nombre de milliardaires et un PIB moyen plus élevé.

```python
# On calcule le nombre de milliardaires par pays
total_billionaires = cleaned_data.groupby('country')['country'].count().reset_index(name='Nombre de milliardaires')

# On calcule le PIB moyen par pays
average_pib = cleaned_data.groupby('country')['gdp_country'].mean().reset_index(name='PIB moyen')

# On fusionne les données avec le merge
data_fusionne = pd.merge(total_billionaires, average_pib, on='country')

# On créé le premier graphique à dispersion
fig = px.scatter(data_fusionne, x='Nombre de milliardaires', y='PIB moyen',
                 title='Relation entre le nombre de milliardaires et le PIB moyen par pays',
                 labels={'Nombre de milliardaires': 'Nombre de milliardaires', 'PIB moyen': 'PIB moyen ($)'},
                 hover_name='country')
fig.update_traces(marker=dict(size=10),
                  selector=dict(mode='markers'))
fig.update_layout(xaxis=dict(type='log'), yaxis=dict(type='log'))
fig.show()

# On calcule le nombre de milliardaires par pays de résidence cette fois-ci vu que l'on s'intéresse aux taxes
total_billionaires = cleaned_data.groupby('countryOfCitizenship')['countryOfCitizenship'].count().reset_index(name='Nombre de milliardaires')

# On prend le taux d'imposition total par pays de résidence
taux_imposition = cleaned_data.groupby('countryOfCitizenship')['total_tax_rate_country'].first().reset_index(name='Taux d\'imposition')

# on fusionne les données avec merge
data_fusionne = pd.merge(total_billionaires, taux_imposition, on='countryOfCitizenship')

# On créé le deuxième graphique à dispersion
fig = px.scatter(data_fusionne, x='Nombre de milliardaires', y='Taux d\'imposition',
                 title='Relation entre le nombre de milliardaires et le taux d\'imposition par pays de résidence',
                 labels={'Nombre de milliardaires': 'Nombre de milliardaires', 'Taux d\'imposition': 'Taux d\'imposition (%)'},
                 hover_name='countryOfCitizenship')
fig.update_traces(marker=dict(size=10),
                  selector=dict(mode='markers'))
fig.update_layout(xaxis=dict(type='log'), yaxis=dict(type='log'))  # Utiliser une échelle logarithmique pour mieux visualiser les données
fig.show()
```

Ce graphique met en évidence la corrélation évoquée juste avant entre le nombre de milliardaires dans un pays et son PIB moyen. Plus précisément, on observe une tendance selon laquelle les pays avec un plus grand nombre de milliardaires tendent également à avoir un PIB moyen plus élevé. Visuellement, les points sur le graphique semblent former une tendance ascendante qui pourrait presque être approximée par une droite.

En revanche, il n'existe pas nécessairement de corrélation directe entre le pays de résidence des milliardaires et le taux de taxes dans ces pays. Les choix de résidence des milliardaires peuvent être influencés par divers facteurs tels que la stabilité politique et les opportunités économiques, indépendamment des politiques fiscales. On peut donc supposer à travers ces résultats que le choix du pays de résidence est indépendant de la fiscalité pour la majorité des milliardaires.

```python
# on prend uniquement les pays avec 50 milliardaires au minimum pour que l'analyse ait du sens
pays_cinquante = cleaned_data['country'].value_counts()[cleaned_data['country'].value_counts() > 50].index

# On filtre les données avec les pays en question
donnees_filtrees = cleaned_data[cleaned_data['country'].isin(pays_cinquante)]

# On calcule le nombre de milliardaires pour chacun de ces pays
billionaires_par_pays = donnees_filtrees.groupby('country').size().reset_index(name='total_billionaires')

# On calcule le nombre de milliardaires par catégorie pour chacun de ces pays
billionaires_par_pays_category = donnees_filtrees.groupby(['country', 'category']).size().reset_index(name='count')

# On fusionne les deux variables que l'on a créé plus haut pour tout avoir dans un seul dataframe
billionaires_par_pays_category = billionaires_par_pays_category.merge(billionaires_par_pays, on='country', how='left')

# on calcule le pourcentage que chaque catégorie représente pour chaque pays
billionaires_par_pays_category['percentage'] = (billionaires_par_pays_category['count'] / billionaires_par_pays_category['total_billionaires']) * 100

# on créé une treemap
fig = px.treemap(billionaires_par_pays_category, path=['country', 'category'], values='count',
                 title='Répartition des milliardaires par catégorie pour les pays avec plus de 50 milliardaires',
                 color='total_billionaires', color_continuous_scale='blues',
                 hover_data={'percentage': True})
fig.show()
```

L'analyse de la répartition des milliardaires par catégorie dans différentes régions met en lumière des tendances distinctes. En Asie, notamment, les secteurs du manufacturing (+ de 20%) et de la santé (environ 15%) se démarquent comme étant les plus propices à la création de richesse extrême. Cette orientation peut refléter la prédominance des industries manufacturières et pharmaceutiques dans cette région dynamique. En revanche, en Europe, les domaines de la finance & investissements (Royaume-Uni et Suisse) ainsi que du fashion and retail (Allemagne et Italie) se distinguent, suggérant une économie axée sur le secteur financier et la mode. Aux États-Unis, la finance & investissement reste un pilier (+ de 25%), mais la technologie émerge également comme une force majeure (17%), probablement due à la présence influente des géants technologiques tels que les GAFAM et l'écosystème dynamique de la Silicon Valley. Ces observations soulignent les nuances régionales dans les opportunités de création de richesse et illustrent l'impact des industries dominantes sur le profil des milliardaires.
